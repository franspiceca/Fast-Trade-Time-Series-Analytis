```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('BTCUSD_2021.csv')

df['date'] = pd.to_datetime(df['date'], unit='s')

df.head()
```

```python
correlation = df['high'].corr(df['low'])
print("Correlation of high and low: ", correlation)
```

```python
correlation = df['open'].corr(df['close'])
print("Correlation of open and close: ", correlation)
```

```python
def plot_df(df, x, y, title="Fast Trade", xlabel='Date', ylabel='Close', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    
plot_df(df, x=df.date, y=df.close, title='Fast Trade')    
```
